```python
import pandas as pd
import numpy as np
```

```python
dados = {
    "Numero_String" : ["1", "2", "3", "4"],
    "Preco_Float" : [1.1, 2.2, 3.3, 4.4],
}

df = pd.DataFrame(dados)
```

```python
print(df)
```

```python
# Converter Numero_String para int
df["Numero_String"] = df["Numero_String"].astype(int)
```

```python
print(df)
```

```python
# Converter Preco_Float para int
df["Preco_Float"] = df["Preco_Float"].astype(int)
```

```python
print(df.info())
print(df)
```

```python
# Valores ausentes
df2 = pd.DataFrame({
    "A" : [1, 2, np.nan],
    "B" : [5, np.nan, np.nan],
    "C" : [7, 8, 9]
})
```

```python
print(df2)
```

```python
#isna()
print(df2.isna())
```

```python
#fillna - Preencher com 0 ontem valores nulos
df_nulos = df2.fillna({"A" : 0})
print(df_nulos)
```

```python
#fillna - Preencher com uma m√©dia de valores
df_media = df2.fillna(df2.mean())
print(df_media)
```

```python
# Remover as linhas com valores ausente
df_drop_linhas = df2.dropna()
print(df_drop_linhas)
```

```python
df_drop_cols = df2.dropna(axis=1)
print(df_drop_cols)
```
