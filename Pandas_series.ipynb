```python
import pandas as pd
```

```python
numeros = [10, 20, 30, 40, 50]

serie1 = pd.Series(numeros)
print(serie1)
```

```python
print(type(serie1))
```

```python
# Acessar um elemento da serie
print(serie1[2])
```

```python
# Criar uma Series com indice personalizado
listas_numeros = [100, 200, 300]
index = ["a", "b", "c"]

serie_personalizada = pd.Series(listas_numeros, index)
print(serie_personalizada)
```

```python
print(serie_personalizada["c"])
```

```python
# Criar uma series a partir de um dicionario
dados = {"primeiro" : 1, "segundo" : 2, "terceiro": 3}
#print(dados)

serie_dict = pd.Series(dados)
```

```python
print(serie_dict)
```

```python
print(serie_dict["terceiro"])
```
