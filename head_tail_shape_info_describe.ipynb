```python
import pandas as pd
```

```python
# Criando um DF
dados = {
    "Nome" : ["João", "Maria", "José", "Ana"],
    "Idade" : [25, 30, 35, 40],
    "Cidade" : ["São Paulo", "Rio de Janeiro", "Belo Horizonte", "Porto Alegre"],
    "Salário" : [5000, 6000, 7000, 8000]
}
```

```python
print(dados)
```

```python
df = pd.DataFrame(dados)
print(df)
```

```python
# head(n) - Retornar as primeiras n linhas do Df (EM BRANCO - padrão 5 primeiras)
print(df.head(9))
```

```python
# tail(n) - Retorna as últimas n linhas do DF
print(df.tail(2))
```

```python
# shape - Retorna quantidade de linhas e colunas
df.shape
```

```python
# info - Fornce um resumo do DF
df.info()
```

```python
# describe() - Retorna estatísticas sobre o DF
print(df.describe())
```
