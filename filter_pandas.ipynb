```python
import pandas as pd
```

```python
# Criar DF
dados = {
    "Nome" : ["João", "Maria", "José", "Ana"],
    "Idade" : [25, 30, 35, 40],
    "Cidade" : ["São Paulo", "São Paulo", "Belo Horizonte", "Porto Alegre"],
    "Pontuação": [85, 90, 95, 100]
}

df = pd.DataFrame(dados)
```

```python
print(df)
```

```python
# Filtrar pessoas com idade maior que 30
filtro_idade = df[df["Idade"] > 30]
```

```python
print(filtro_idade)
```

```python
# Cidade São Paulo e pontuação > 80
filtro_cid_pont = df[(df["Cidade"] == "São Paulo") & (df["Pontuação"] > 80)]
```

```python
print(filtro_cid_pont)
```

```python
# Cidade Fortleza ou Campina Grande
filtro_cidade = df[(df["Cidade"] == "Fortaleza") | (df["Cidade"] == "Campina Grande")]
print(filtro_cidade)
```

```python
# Comando isin()
filtro_isin = df[df["Cidade"].isin(["Goiania", "Campinas", "São Paulo"])]
print(filtro_isin)
```
