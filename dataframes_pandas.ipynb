```python
import pandas as pd
```

```python
# Criando nosso primeiro DataFrame a partir de um dicionário
dados = {
    "Nome" : ["Alice", "Eduardo", "Carlos", "Davi"],
    "Idade" : [25, 30, 19, 27],
    "Cidade" : ["São Paulo", "Rio de Janeiro", "Belo Horizonte", "Curitiba"]
}

print(dados)
```

```python
df = pd.DataFrame(dados)
print(df)
```

```python
# Acessando uma coluna especifica do DF
print(df.Nome)
```

```python
print(df["Cidade"])
```

```python
# Criar um DF a partir de uma lista de dicionarios
lista_dict = [
    {"Nome" : "Alice", "Idade" : 25, "Cidade" : "São Paulo"},
    {"Nome" : "Eduardo", "Idade" : 30, "Cidade" : "Rio de Janeiro"},
    {"Nome" : "Carlos", "Idade" : 19, "Cidade" : "Belo Horizonte"},
]

df2 = pd.DataFrame(lista_dict)
```

```python
print(df2)
```

```python
print(df2.Nome)
```
