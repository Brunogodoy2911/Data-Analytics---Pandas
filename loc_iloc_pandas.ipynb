```python
import pandas as pd
```

```python
# Criar o DF
dados = {
    "Produto" : ["A", "B", "C", "D", "E"],
    "Preco" : [10, 20, 30, 40, 50],
    "Estoque" : [100, 50, 75, 20, 30]
}

index = ["item1", "item2", "item3", "item4", "item5"]

df = pd.DataFrame(dados, index)
```

```python
print(df)
```

```python
# Visualizar colunas específicas
df.Produto
df["Estoque"]
```

```python
# Selecionar múltiplas colunas
print(df[["Produto", "Estoque"]])
```

```python
# Comando loc - Seleção do item3
print(df.loc["item3"])
```

```python
# Comando loc para multiplas colunas
print(df.loc[["item1", "item5"], ["Preco", "Estoque"]])
```

```python
# Comando iloc para buscar o item3

print(df.iloc[2])
```

```python
# Comando iloc para linhas e colunas
print(df.iloc[0:2, 1:3])
```
